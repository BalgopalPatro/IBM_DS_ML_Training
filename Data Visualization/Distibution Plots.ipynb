```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
type(tips)
```

## Distplot

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['size'])
```

```python
sns.distplot(tips['total_bill'],bins=20,kde=False,rug=True)
```

```python
ax = sns.distplot(tips['total_bill'], rug=True, rug_kws={"color": "y"},
                  kde_kws={"color": "k", "lw": 2, "label": "KDE"},
                  hist_kws={"histtype": "step", "linewidth": 1,
                  "alpha": 1, "color": "g"})
```

```python
sns.distplot([1,2,3,4,5,5,5,5],rug=True, rug_kws={"color": "r"})
```
