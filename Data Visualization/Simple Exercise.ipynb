```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('../pandas/Car_sales.csv')
data.head()
```

```python
print("Shape Of data",data.shape)
```

```python
pivotData = data.pivot_table(values='Sales in thousands',columns='Latest Launch',index='Manufacturer')
```

```python
print("Shape Of data",pivotData.shape)
```

```python
pivotData.head()
```

## Treating nan Values

```python
pivotData.fillna(0,inplace=True)
```

```python
pivotData.head()
```

```python
fig = plt.figure(figsize=(20,10))
sns.heatmap(pivotData,cmap='RdBu')
```
