```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(0,10)
y = x ** 4
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title('Plot Title')
plt.show()
```

```python
plt.plot(y,x,'r*-')
plt.show()
```

```python
c = np.arange(-5*np.pi,5*np.pi)
plt.subplot(211)
plt.plot(c,np.sin(c),'b')
plt.subplot(212)
plt.plot(c,np.cos(c),'r')
plt.show()
```

```python
xx=np.linspace(-10,10,150)
cos=np.cos(xx)
sin=np.sin(xx)
plt.subplot(211)
plt.plot(xx,cos,'b')
plt.subplot(212)
plt.plot(xx,sin,'r')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.subplot(122)
plt.plot(y,x)
plt.show()
```

```python
xx = np.linspace(-np.pi,np.pi,20)
sin = np.sinh(xx)
cos = np.cosh(xx)
plt.subplot(121)
plt.plot(xx,sin)
plt.subplot(122)
plt.plot(xx,cos)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(y,x,'b*-')
ax2 = fig.add_axes([0.5,0.1,.4,.5])
ax2.plot(x,y,'r+-')
plt.show()
```
