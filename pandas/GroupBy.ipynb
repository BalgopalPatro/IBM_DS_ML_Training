```python
import pandas as pd
```

```python
data = {
    'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
    'Person':['Bg','Balgopal','Reshuu','Reshma','Papu','Subham'],
    'Sales':[102,129,192,179,223,132]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
grp_by_com = df.groupby('Company')
```

```python
grp_by_com.mean()
```

```python
grp_by_com.count()
```

```python
grp_by_com.describe()
```

```python
grp_by_com.describe().transpose()
```
